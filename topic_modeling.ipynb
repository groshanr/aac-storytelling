```python
from datasets import load_dataset
import pandas as pd
import datasets
from datasets import Dataset, DatasetDict
from fastopic import FASTopic
from topmost import Preprocess

ds = load_dataset("FareedKhan/1k_stories_100_genre")
```

```python
train_ds = ds['train']
```

```python
train_df = train_ds.to_pandas()
```

```python
train_df.drop(columns=['title', 'id'], inplace=True)
```

```python
train_df
```

```python
scifi_df = train_df[train_df['genre'] == 'Science Fiction']
```

```python
scifi_df.head()
```

```python
scifi_ds = Dataset.from_pandas(scifi_df)
```

```python
adventure_df = train_df[train_df['genre'] == 'Adventure']
```

```python
adventure_ds = Dataset.from_pandas(adventure_df)
```

```python
mystery_df = train_df[train_df['genre'] == 'Mystery']
```

```python
mystery_ds = Dataset.from_pandas(mystery_df)
```

```python
fantasy_df = train_df[train_df['genre'] == 'Fantasy']
```

```python
fantasy_ds = Dataset.from_pandas(fantasy_df)
```

```python
horror_df = train_df[train_df['genre'] == 'Horror']
```

```python
horror_ds = Dataset.from_pandas(horror_df)
```

```python
# Prepare your dataset.
adventure_docs = adventure_ds['story']

# Preprocess the dataset. This step tokenizes docs, removes stopwords, and sets max vocabulary size, etc.
# preprocess = Preprocess(vocab_size=your_vocab_size, tokenizer=your_tokenizer, stopwords=your_stopwords_set)
preprocess = Preprocess()

model = FASTopic(10, preprocess)
adventure_top_words, adventure_doc_topic_dist = model.fit_transform(adventure_docs)
```
